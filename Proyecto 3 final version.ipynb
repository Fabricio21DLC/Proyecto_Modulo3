## Objetivos:

### General:
- Comparar 3 distintas acciones de la bolsa de valores que se encuentren cercanamente en el mismo rubro de negocio.

### Especificos:
- Comparar el valor de las 3 acciones y ver en cual le convendría a cada quien invertir.
- Obtener los máximos y mínimos de cada acción para poder obtener la mayor ganancia que se podía obtener en cierto período de tiempo.


## Modelo a representar:

- Debemos crear un código que nos permita hacer un barrido para partir los distintos datos que insertamos usando pandas,
esto permitira que podamos crear otra función que nos regrese cuáles son los 3 mejores máximos y mínimos que se podían
usar para obtener una ganacia.

### Acciones a evaluar: 

- Motorola
- Apple
- Sony

## Simulaciónes y gráficas: 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Motorola

```python
moto="MSI.csv"
```

```python
df_moto =pd.read_csv(moto)
```

```python
y=df_moto["Adj Close"]
x=df_moto["Tiempo"]
adjclose = y.values
dias = x.values
```

```python
plt.figure(figsize=(6,4))
plt.plot(dias,adjclose)
plt.xlabel("Dias", fontsize=18)
plt.ylabel("AdjClose", fontsize=18)
plt.grid()

plt.show()
```

## Gráfica dividida cada 2 meses (Al ser un márgen tan extenso contiene inconsistencias en los máximos y mínimos.)

```python
i = 0
diasTotales = 1260
secciones = 21
rango = int(diasTotales/secciones)
valores = []
minimos = []
maximos = []
diaMinimos = []
diaMaximos = []

while (i<diasTotales):
    valores = adjclose[i:i+rango]
    
    minimos.append(min(valores))
    maximos.append(max(valores))
    
    diaMinimos.append(np.argmin(valores)+i)
    diaMaximos.append(np.argmax(valores)+i)
    
    i = i + rango;

plt.figure(figsize=(10, 10))

plt.xlabel("Dias", fontsize=18)
plt.ylabel("AdjClose", fontsize=18)

for i in range(secciones):
    plt.scatter(diaMinimos[i], minimos[i], c="r")
    plt.scatter(diaMaximos[i], maximos[i], c="b")
    
plt.plot(dias,adjclose)
plt.grid()
plt.show ()
```

```python
diferencias = []

for i in range(secciones):
    diferencias.append(maximos[i] - minimos[i])

difDesordenadas = list(diferencias)
diferencias.sort()
print(diferencias[-3:])
plt.figure(figsize=(8,6))

for i in range(1,4):
    posicion = difDesordenadas.index(diferencias[-(i)])
    plt.scatter(diaMinimos[posicion],minimos[posicion], c="r")
    plt.scatter(diaMaximos[posicion],maximos[posicion], c="b")
    print(diaMinimos[posicion],minimos[posicion])
    print(diaMaximos[posicion],maximos[posicion])


plt.xlabel('Dias', fontsize=18)
plt.ylabel("AdjClose", fontsize=18)

plt.plot(dias, adjclose)
plt.grid()
plt.show()
```
